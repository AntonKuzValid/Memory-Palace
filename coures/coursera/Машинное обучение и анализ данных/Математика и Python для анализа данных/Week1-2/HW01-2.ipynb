Рассмотрим сложную математическую функцию на отрезке [1, 15]:

f(x) = sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2)

```python
import numpy as np
import scipy.linalg as linalg

def funcF(x):
   return np.sin(x/5)*np.exp(x/10)+5*np.exp(-x/2)

def initX(x,n):
    result=np.zeros((len(x), n+1))
    for i in range(n+1):
        result[:, i] = x ** i
    return result

def main(lis,n):
    arr = np.array(lis)
    x = initX(arr,n)
    b = funcF(arr)
    w=linalg.solve(x,b)
    allx= initX(np.arange(1,16),n)
    plot(allx[:,1], allx.dot(w))
    return w
```
